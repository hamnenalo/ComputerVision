```python
import cv2
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
img = cv2.imread("20220227-A7401455_TP_V4.jpg")
```

```python
img_r = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.figure(1, (12., 12.))
plt.imshow(img_r)
```

```python
resize_img = cv2.resize(img, (380, 250))
```

```python
img_r = cv2.cvtColor(resize_img, cv2.COLOR_BGR2RGB)
fig = plt.figure(1, (12., 12.))
plt.imshow(img_r)
```

```python
resize_img = cv2.resize(img, (190, 125))
```

```python
img_r = cv2.cvtColor(resize_img, cv2.COLOR_BGR2RGB)
fig = plt.figure(1, (12., 12.))
plt.imshow(img_r)
```
